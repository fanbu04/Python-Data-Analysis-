## Part 2: Creating Data Munging Functions
Create one function to perform each of the following tasks: 
#### 1. Remove symbols from column headers
    - This function should take in a dataframe and a symbol to remove as arguments
    - It should remove the symbol from all column names

#### 2. Replace missing numeric values with the column mean
    - This function should take in a dataframe and column names as arguments
    - For the specified column, it should replace any null values with the mean for the column

Write those functions below: 

```python
adult.columns = adult.columns.str.replace("$",'').str.replace("!",'')
adult.head()
```

```python
adult["age"].fillna(adult["age"].mean(),inplace= True)
#adult["workclass"].fillna(adult["workclass"].mean(),inplace= True)
adult["fnlwgt"].fillna(adult["fnlwgt"].mean(),inplace= True)
#adult["education"].fillna(adult["education"].mean(),inplace= True)
adult["education_num"].fillna(adult["education_num"].mean(),inplace= True)
adult["capital_gain"].fillna(adult["capital_gain"].mean(),inplace= True)
adult["capital_loss"].fillna(adult["capital_loss"].mean(),inplace= True)
adult["hours_per_week"].fillna(adult["hours_per_week"].mean(),inplace= True)
```

## Part 3: Apply the functions to the adult dataset 
Apply each of your functions to the adult dataset. Check to make sure they worked properly. 

```python
def remove_sym (df,symbol):
    df.columns = df.columns.str.replace(symbol,'')
    return df
```

```python
adult_remove = remove_sym(adult,"$")
adult_remove = remove_sym(adult,"!")

adult_remove.columns
```

```python
adult.dtypes
```

```python
def replace(df):
    for col in df.columns:
        if df[col].dtypes == 'int64'or df[col].dtypes == 'float64':
            df[col].fillna(df[col].mean(),inplace= True)
        
    return df
    
    
```

```python
adult_replace = replace(adult_remove)
adult_replace.isnull()
```

## Part 4: Grouping
Group the adult dataset by education and calculate the mean age for each group. 

```python
adult_classes = adult_replace.groupby("education")
list(adult_classes)
```

```python
adult_classes["age"].mean()
```
